# Deep Layer
### testing generator model (from Generative Adversarial Network) on a test image

##### importing libraries

```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from services.img import add_padding, remove_padding, normalize_irr, preprocess_images_with_padding
```

##### define test image name and load from data/test folder
###### rename file_name with the relative path of the image in the test folder rgb and irr (jpg format)

```python
file_name = '05'
rgb_image_path = f'../data/test/rgb/{file_name}.jpg'
irr_image_path = f'../data/test/irr/{file_name}.jpg'
```

```python
rgb_image = cv2.imread(rgb_image_path)
rgb_image_padded = add_padding(rgb_image, (512, 512))
irr_image = cv2.imread(irr_image_path)
irr_image_padded = add_padding(irr_image, (512, 512))
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title("RGB Image")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(irr_image, cv2.IMREAD_GRAYSCALE))
plt.title("IRR Layer Ground Truth")
plt.show()
```

#### load generator model and check parameters and layer

```python
generator = tf.keras.models.load_model(f'../models/generator.keras')
generator.summary()
```

#### generate underpainting with the model and plot the results

```python
#rgb_image = np.expand_dims(rgb_image, axis = 0)
underpainting = generator.predict(
    np.expand_dims(preprocess_images_with_padding(rgb_image_padded, rgb_image_padded, 512)[0],
    axis = 0)
)
```

```python
# plot example images
plt.figure(figsize=(15, 8))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(rgb_image_padded, cv2.COLOR_BGR2RGB))
plt.title("RGB Image")
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(irr_image_padded, cv2.IMREAD_GRAYSCALE))
plt.title("IR Layer Ground Truth")
plt.subplot(1, 3, 3)
plt.imshow(underpainting[0], cmap='gray')
plt.title("IRR Layer predicted")
plt.show()
```

```python
rgb_image_cropped = remove_padding(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB), rgb_image_padded)
irr_image_cropped = remove_padding(rgb_image, irr_image_padded)

# transform original IRR in grayscale, set size, trasform to float and normalize from 0 to 1
gray_image = cv2.cvtColor(irr_image, cv2.COLOR_BGR2GRAY)
gray_image = normalize_irr(gray_image)
gray_image_float32 = gray_image.astype(np.float32)
gray_image_float32 = gray_image_float32 / 255.0

gray_image_float32 = add_padding(gray_image_float32,(512, 512))
gray_image_float32 = remove_padding(irr_image, gray_image_float32)

# normalize the predicted IRR
underpainting_no_pading = remove_padding(irr_image, underpainting[0])
underpainting_normalized = normalize_irr(underpainting_no_pading)
underpainting_normalized = underpainting_normalized /255.0
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.imshow(underpainting.squeeze(), cmap='gray')
plt.title("Predicted")
plt.subplot(1, 2, 2)
plt.imshow(underpainting_normalized, cmap='gray')
plt.title("De-padded & normalized")
plt.show
```

```python
print("---"*10)
print("Original IRR image")
print(irr_image_cropped.shape)
print("---"*10)
print("Original IRR image in grayscale")
print(gray_image_float32.shape)
print("---"*10)
print("Predicted IRR image")
print(underpainting_normalized.shape)
print("---"*30)
print("Original IRR image in grayscale")
print(f"Max of gray image: {np.max(gray_image_float32):.5f}")
print(f"Min of gray image: {np.min(gray_image_float32):.5f}")
print("Predicted IRR image")
print(f"Max of predicted image: {np.max(underpainting_normalized):.5f}")
print(f"Min of predicted image: {np.min(underpainting_normalized):.5f}")
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(2, 4, 1)
plt.imshow(cv2.cvtColor(rgb_image_cropped, cv2.COLOR_BGR2RGB))
plt.title("Original RGB")
plt.subplot(2, 4, 2)
plt.imshow(irr_image_cropped, cmap='gray')
plt.title("Original IRR")
plt.subplot(2, 4, 6)
plt.hist((irr_image_cropped/255.0).ravel(), bins=256)
plt.subplot(2, 4, 3)
plt.imshow(gray_image_float32,cmap='gray')
plt.title("IRR in grayscale, normalized")
plt.subplot(2, 4, 7)
plt.hist(gray_image_float32.ravel(), bins=256)
plt.subplot(2, 4, 4)
plt.imshow(underpainting_normalized, cmap='gray')
plt.title("Predicted IRR, normalized")
plt.subplot(2, 4, 8)
plt.hist(underpainting_normalized.ravel(), bins=256)
plt.show()
```

#### Create the difference map
This is useful for visualizing areas where the model underestimates (red, positive values) or overestimates (blue, negative values) the values.

```python
if len(underpainting_normalized.shape) > 2:
    underpainting_normalized = underpainting_normalized.squeeze()
diff = gray_image_float32 - underpainting_normalized
```

#### Global Error Evaluation

```python
print(f"Max difference: {np.max(diff):.5f} (understimation of prediction)")
print(f"Min difference: {np.min(diff):.5f} (overstimation of prediction)")
mse = (diff**2).mean()
rmse = np.sqrt(mse)
print(f"MSE: {mse:.5f}")
print(f"RMSE: {rmse:.5f}")
print(f"Non zero values: {np.count_nonzero(diff)} over {diff.size}")
```

#### Evaluation of underestimation errors (positive values)

```python
diff_sup = diff[diff > 0]
print(f"Max : {np.max(diff_sup):.5f}")
print(f"Min : {np.min(diff_sup):.5f}")
mse = (diff_sup**2).mean()
rmse = np.sqrt(mse)
print(f"MSE: {mse:.5f}")
print(f"RMSE: {rmse:.5f}")
print(f"Values greater than zero: {np.count_nonzero(diff_sup)} over {diff.size}")
```

#### Evaluation of overestimation errors (negative values)

```python
diff_inf = diff[diff < 0]
print(f"Max : {np.max(diff_inf):.5f}")
print(f"Min : {np.min(diff_inf):.5f}")
mse = (diff_inf**2).mean()
rmse = np.sqrt(mse)
print(f"MSE: {mse:.5f}")
print(f"RMSE: {rmse:.5f}")
print(f"Values less than zero: {np.count_nonzero(diff_inf)} over {diff.size}")
```

#### Near zero values

```python
diff_max = 0.01
diff_min = -0.01
near_zero = diff[(diff<diff_max) & (diff>diff_min)]
print(f"Max : {np.max(near_zero):.5f}")
print(f"Min : {np.min(near_zero):.5f}")
mse = (near_zero**2).mean()
rmse = np.sqrt(mse)
print(f"MSE: {mse:.5f}")
print(f"RMSE: {rmse:.5f}")
print(f"Near zero values: {np.count_nonzero(near_zero)} over {diff.size}")
```

#### Plot difference image and distribution

```python
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.imshow(diff, cmap='seismic', vmin=-1, vmax=1)
plt.title("IR Layer difference (original - predicted)")
plt.colorbar(label="difference")
plt.subplot(1, 2, 2)
plt.hist(diff_inf, bins=256, range=[-1,1], color='#00008b')
plt.hist(diff_sup, bins=256, range=[-1,1], color='#8b0000')
plt.hist(near_zero, bins=256, range=[-1,1], color='#ccc')
plt.show()
```
